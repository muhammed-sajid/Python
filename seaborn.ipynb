Count plot of survivors vs non-survivors

Count plot of survivors vs sex

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as  pd

data1 = pd.read_csv('train.csv')

plt.figure(figsize=(8, 5))
sns.countplot(data=data1, x='Sex', hue='Survived', palette='pastel')
plt.title("Count of Survivors vs Sex", fontsize=12)
plt.xlabel("Sex", fontsize=10)
plt.ylabel("Count", fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=data1, x='Survived',hue='Survived', palette='pastel')
plt.title('Count of Survivors vs Non-Survivors', fontsize=14)
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

 Create a boxplot of fares for each passenger class

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('train.csv')

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Pclass', y='Fare', palette='coolwarm')
plt.title('Boxplot of Fares by Passenger Class ',fontsize=14)
plt.xlabel('Passenger class(Pclass)',fontsize=12)
plt.ylabel('Fare', fontsize=12)
plt.show()
```
