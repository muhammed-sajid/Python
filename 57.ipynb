```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)

#Plot a histogram of passenger ages
ages = data['Age'].dropna()

plt.figure(figsize=(8, 5))
plt.hist(ages, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of passengers Ages', fontsize=12)
plt.xlabel('Age', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7 )
plt.show()
```

```python
fare_by_class = data.groupby('Pclass')['Fare'].mean()

#Create a line plot for fare by passenger class
plt.figure(figsize=(8, 5))
plt.plot(fare_by_class.index, fare_by_class.values, marker='o', linestyle='--', color='orange', label='Average fare')
plt.title('Average Fare by passengers class',fontsize=14)
plt.xlabel('Passengers Class', fontsize=12)
plt.ylabel('Average Fare', fontsize=12)
plt.xticks(fare_by_class.index, ['Class 1', 'Class 2', 'Class 3'], fontsize=10 )
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
