```python
from dataclasses import dataclass
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


data = pd.read_csv('train.csv')

# Data Cleaning: Ensure 'age' and 'fare' have no missing values
data = data[['Age','Fare','Pclass']].dropna()

sns.set_style('whitegrid')

# Creating the scatter plot
plt.figure(figsize=(10, 6))

scatter = sns.scatterplot(data=data, x='Age', y='Fare', hue='Pclass', palette='viridis', s=100, edgecolor='w')

# Adding plot enhancements
plt.title('Scatterplot Age vs Fare by passenger Class', fontsize=16, weight='bold')
plt.xlabel('Age',fontsize=12)
plt.ylabel('Fare',fontsize=12)
plt.legend(title='Passenger Class', loc='upper right')
plt.grid(visible=True, linestyle='--', alpha=0.7)

plt.show()
```
